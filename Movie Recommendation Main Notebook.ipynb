We need to clean the genres, keywords, cast and crew columns from the id variables and convert them neatly to only get the names and actual values. We define a method to do this in a reusable way

```python
import ast

def cleanCol(colName):
    result = []
    for i in ast.literal_eval(colName): # Since col has data that are 'stringified', I chose ast lib to convert it to an object (parse)
        result.append(i['name']) 
    return result
```

```python
df_movies['genres'] = df_movies['genres'].apply(cleanCol)
```

```python
df_movies['keywords'] = df_movies['keywords'].apply(cleanCol)
```

```python
df_movies['cast'] = df_movies['cast'].apply(cleanCol)
```

```python
df_movies['cast'] = df_movies['cast'].apply(lambda x:x[0:3]) # slice from index 0 to 3, i.e, return first 3
```

We will discard all crew members from dataset except the director. Therefore, we create a method to do this for us. 
Since we do not want anyone else from the crew list, therefore a custom new method was needed.

```python
def getDirector(arr):
    result = []
    for i in ast.literal_eval(arr):
        if i['job'] == 'Director':
            result.append(i['name'])
    return result
```

```python
df_movies['crew'] = df_movies['crew'].apply(getDirector)
```

Observe that there are spaces in most of the individual elements for keywords, cast & crew that can cause problems for us in calculations and tokenization later. Therefore, we have to get rid of them except for overview since we will create strings from that.


```python
def cleanSpaces(colData):
    result = []
    for i in colData:
        result.append(i.replace(" ",""))
    return result
```

## Data Processing - Vectorization 

We have to convert every tag (text) to a vector. Therefore, every movie is a vector in other words.
Also called Text Vectorization. We will use Bag of Words  technique (TF IDF, Word2Vec are alternatives)

Every movie has a tag. Combine all tags.
If 100 words, you will get 1k words.
Then the words that are used the most .
I have to calculate how many times that word. 
Extract words that are the most occurring from the large combined text. 
Now take the tag of every movie again. Lets compare how many times those words we got before appear in the tag row of every movie. For example: tomcruise in movie 1 appears 5 time, etc. tomcruise in movie 6 appears 2 times.
This repeated for all 5000 movies. Dimension of this table will be 5000xnoOfTotalWords. 
Therefore, every movie will now be a vector.
Finally, what I will do is compare distances of each of these vectors and the closest one will appear as a recommended movie. Using cosine similarity for distances of vectors

Note: I will remove stop words in my vectorization process for more accuracy.

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
cv = CountVectorizer(max_features=10000,stop_words='english')
```

```python
cv.fit_transform(cleanedData['tags'])
```

```python
vector = cv.fit_transform(cleanedData['tags']).toarray() # convert it to array since it returns a sparse matrix
```

```python
vector
```

0 means the word did not appear in that movie. I expect many 0s in the matrices since there will be very less words from the global total words that will appear in each movie.

```python
cv.get_feature_names()
```

Observe how these feature names look like right now. Arranged alphabetically
There is one problem with them. Words like actions,actionhero, action are repeated as separate features.
Therefore, I have to go back to my original cleanedData frame and perform stemming on them to normalise all word features

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
# Function to stem all strings in the tags column
def stemItem(str):
    result = []
    for i in str.split():
        result.append(ps.stem(i))
        
    string = " ".join(result)
    return string
```

```python
cleanedData['tags'] = cleanedData['tags'].apply(stemItem)
```

```python
# Have to execute this command again to create the vector again
vector = cv.fit_transform(cleanedData['tags']).toarray() # convert it to array since it returns a sparse matrix
```

```python
vector
```

```python
cv.get_feature_names()
```

Now we are bound to get more accurate results since the stemming process has normalised the different versions of a word as one.

```python
similarity = cosine_similarity(vector)
```

```python
similarity[0]
```

Lets take first element of the similarity result
The similarity for movie 1 shows us similarity between movie 1 and every other movie (higher distance = similarity closer to 0 and vice versa). So for itself it will be 1 since it is the MOST similar of every movie with the other movies.

```python
# Find index of a movie in the dataframe
cleanedData[cleanedData['title'] == 'Batman'].index[0]
```

```python
sorted(similarity[1362],reverse=True)
```

Problem with this sorting is that we lose the indexes at which the movie numbers are saved in our data frame. 
Therefore, we need a way to record which index do they originally exist at

```python
enumerate(similarity[0])
```

```python
list(enumerate(similarity[0])) # Saves index of movie with its similarity. 
```

```python
sorted(list(enumerate(similarity[0])), reverse=True)
```

Unfortunately sorting is done on basis of the movie position, not the similarities. Therefore, I make a slight change

```python
# x[1] chooses the second element of the first similarity element. Second element is the similarity
sorted(list(enumerate(similarity[0])), reverse=True, key = lambda x: x[1]) 
```

```python
#  Lets create a method using our above findings
def recommend(movieName):
    idx = cleanedData[cleanedData['title'] == movieName].index[0]
    distance = sorted(list(enumerate(similarity[idx])), reverse=True, key = lambda x: x[1])
    for i in distance[0:6]: # Only first 6 recommended movies
        print(cleanedData.iloc[i[0]].title) # iloc gets location of the movie in the dataframe using the index in distance array's first element. E.g, (1216, 0.2412141)
    
```

```python
# Testing to see what I get from the function
recommend('Avatar')
```
